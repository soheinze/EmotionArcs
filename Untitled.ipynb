```python
test_df = pd.read_csv('soundtracksData.csv')
```

```python
test_df.columns
```

```python
import json
```

```python
test_df['valence_list'] = test_df['valence_list'].apply(json.loads)
```

```python
test_df['energy_list'] = test_df['energy_list'].apply(json.loads)
```

```python
bool_series = test_df['genres'].str.contains("Family")
filtered_df = test_df[bool_series]
filtered_df.reset_index(drop=True,inplace=True)
```

```python
filtered_df['valence_list']
```

```python
x_list = []
y_list = []

for f_list in filtered_df['valence_list']:
    for i,value in enumerate(f_list):
        x_list.append(i/len(f_list))
        y_list.append(value)
```

```python
px.scatter(x=x_list,y=y_list)
```

```python
from tslearn.utils import to_time_series_dataset
X = to_time_series_dataset([[1, 2, 3, 4], [1, 2, 3], [2, 5, 6, 7, 8, 9]])
```

```python
X
```

```python
from tslearn.clustering import TimeSeriesKMeans
km = TimeSeriesKMeans(n_clusters=2, metric="dtw")
labels = km.fit_predict(X)
km_bis = TimeSeriesKMeans(n_clusters=2, metric="softdtw")
labels_bis = km_bis.fit_predict(X)
```
